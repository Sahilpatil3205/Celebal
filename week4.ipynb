```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
sns.set(style="whitegrid")
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
df.describe(include='all')
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]
print(missing)

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="YlOrRd")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Survived', hue='Survived', data=df, palette='Set2', legend=False)
plt.title("Survival Count")
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Pclass', data=df, palette='Set1', legend=False)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='teal')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', hue='Pclass', data=df, palette='Set3', dodge=False, legend=False)
plt.title("Age by Passenger Class")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(y='Fare', data=df)
plt.title("Fare Distribution and Outliers")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df, palette='husl')
plt.title("Survival by Port of Embarkation")
plt.show()
```
